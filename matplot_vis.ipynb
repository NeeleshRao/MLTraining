```python
# the kx can be changed to many parameters like bH for blue hexagon dot
plt.plot([[4, 3, 5, 3, 8], [9, 8, 7, 6, 5], [2, 4, 6, 8, 10]], "kx-")
# Here, we are passing 2d matrix, and they are indexed as 0th, 1st and 2nd column at x=0, ,1, 2
plt.xlabel("Some values")
plt.ylabel("Some data")
# plt.show()
```

```python
# c=[50, 100, 105, 200, 2500]
some=np.array([0, 1, 2, 3, 4])
colors=np.array(['red', 'green', 'blue', 'yellow', 'black'])
plt.scatter([9, 8, 7, 6, 5], [5, 4, 3, 2, 1], s=[5, 10, 15, 20, 25], c=colors[some])
```
