```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Better to drop the user id column as it is useless
data = pd.read_csv("Social_Network_Ads.csv")

target = data["Purchased"]
data = data.drop("Purchased", axis=1)
data = data.drop("User ID", axis=1)
data["Gender"] = data["Gender"].apply(lambda x: 1.0 if x=="Male" else 0.0 )
print(target)
print(data)

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2)

svm = SVC(kernel="linear")
svm.fit(X_train, y_train)

out = svm.predict(X_test)
print(svm.score(X_train, y_train)*100 ,"%")
print(accuracy_score(y_test, out)*100, "%")
```
